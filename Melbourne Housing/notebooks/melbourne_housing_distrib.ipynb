```python
import pandas as pd
import seaborn as sns
sns.set(style = 'darkgrid', font_scale = 1.2)
```

```python
df = pd.read_csv('../data/housing/melb_data.csv', usecols=['Regionname', 'Type', 'Rooms', 'Distance',
                                                          'Price'])
df.head()
```

```python
sns.displot(
    data = df,
    x = 'Price',
    kind = 'hist',
    aspect=1.4
)
```

```python
sns.displot(
    data = df,
    x = 'Price',
    kind = 'hist',
    aspect=1.4,
    log_scale=10
)
```

```python
sns.displot(
    data = df,
    x = 'Price',
    kind = 'hist',
    aspect = 1.4,
    log_scale=10,
    bins = 20
)
```

```python
sns.displot(
    data = df,
    x = 'Price',
    hue = 'Type',
    kind = 'hist',
    aspect=1.4,
    log_scale=10,
    bins = 20
)
```

```python
sns.displot(
    data = df,
    x = 'Price',
    col = 'Type',
    kind = 'hist',
    aspect=1.4,
    log_scale=10,
    bins = 20
)
```

```python
sns.displot(
    data = df,
    x = 'Price',
    y = 'Distance',
    col = 'Type',
    kind = 'hist',
    height = 5,
    aspect=1.2,
    log_scale=(10,0),
    bins = 20
)
```

```python
sns.displot(
    data = df,
    x = 'Price',
    kind = 'kde',
    aspect = 1.4,
    log_scale = 10
)
```

```python
sns.displot(
    data = df,
    x = 'Price',
    hue = 'Regionname',
    kind = 'kde',
    height =6,
    aspect=1.4,
    log_scale=10
)
```

```python
sns.displot(
    data = df,
    x = 'Distance',
    kind = 'ecdf',
    height = 6,
    aspect = 1.4,
    stat = "count"
)
```

```python
sns.displot(
    data = df,
    x = 'Distance',
    kind = 'ecdf',
    hue = 'Type',
    height = 6,
    aspect = 1.4,
    stat = 'count'
)
```
